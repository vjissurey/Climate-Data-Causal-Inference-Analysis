## Time series Analysis 

```python
import pandas as pd

enso_ts = pd.read_csv('./enso_ond(1).csv')
jet_ts = pd.read_csv('./jet_ond(1).csv')
spv_ts = pd.read_csv('./spv_ond(1).csv')

# Set 'year' as index
enso_ts.set_index('year', inplace=True)
jet_ts.set_index('year', inplace=True)
spv_ts.set_index('year', inplace=True)

# Plotting each time series
plt.figure(figsize=(10, 5))
plt.plot(enso_ts.index, enso_ts['enso'], label='ENSO')
plt.title('ENSO Time Series')
plt.xlabel('Year')
plt.ylabel('ENSO Index')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(jet_ts.index, jet_ts['jet'], label='Jet', color='orange')
plt.title('Jet Time Series')
plt.xlabel('Year')
plt.ylabel('Jet Index')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(spv_ts.index, spv_ts['spv'], label='SPV', color='green')
plt.title('SPV Time Series')
plt.xlabel('Year')
plt.ylabel('SPV Index')
plt.legend()
plt.show()
```
